### Importing the packages

```python
import time
import cv2
import os
import matplotlib.pyplot as plt
```

### Open an image

```python
img = cv2.imread("./Test/tt20.png")
plt.imshow(img[:,:,::-1])
plt.show()
```

### Cropping the image to show the number plate

```python
# Cropout OpenCV logo
img1 = img[200:,95:150]
plt.imshow(img1[:,:,::-1])
plt.show()
```

### Using the EDSR X4 Model

```python
super_resolution = cv2.dnn_superres.DnnSuperResImpl_create()
```

```python
path = "./Models/EDSR_x4.pb"
```

```python
super_resolution.readModel(path)
super_resolution.setModel("edsr",4)
```

```python
result = super_resolution.upsample(img1)

# Resized image
resized = cv2.resize(img1, dsize = None, fx=4,fy=4)

# Displaying
plt.figure(figsize=(12,8))
plt.subplot(1,3,1)
# Original image
plt.imshow(img1[:,:,::-1])
plt.subplot(1,3,2)
# SR upscaled
plt.imshow(result[:,:,::-1])
plt.subplot(1,3,3)
# OpenCV upscaled
plt.imshow(resized[:,:,::-1])
plt.show()
```
